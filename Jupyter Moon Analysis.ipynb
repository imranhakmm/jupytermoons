## Estimating the mass of Jupyter

Kepler's law states that the square of a planet's orbital period is proportional to the cube of the length of the semi-major axis  of its rbito.
  
- From the formula we see that the constant of proportionality for this relationship is {4*pi^2/GM}, where G is the gravitational constant and M is the mass of the orbitted body 
- We will apply this theorem to finding M, where M is the mass of Jupiter. This makes sense since we have the values of orbital periods and distances of the moons of Jupiter. For simplicity, we will use distance between the moon and jupyter as the proxy for semi-major axis length.
- Since there is only 1 constant of proportionality, it makes sense to model this relationship with Linear regression. However, we will verify this later.

We follow these steps:

1. Prepare Data : Adjust units for period and distance, and calculate T_squared and a_cubed.
2. Verify Linear Relationship : We start by examining the linear relationship between the square of the moons' orbital periods(T_squared) and the cube of their semi-major axes(a_cubed)
3. Train Linear Regression Model : We train a linear regression model on this dataset.
4. Calculate and Compare Jupiter's Mass : Using the linear model, we estimate Jupiter's mass and we compare our estimated value with the known theoretical value of Jupiter's mass.

To assist this process, we will use another module called "MassModelling" that suppport additonal functionalities for our tass.
