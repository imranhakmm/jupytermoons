# Jupiter's Moons Data Analysis

In this Jupyter Notebook, we will explore and analyze data related to the moons of Jupiter. We will utilize the `Moons` class from the `jupiter.py` module, which provides various functionalities for loading, processing, and visualizing the data. Our goal is to gain insights into the characteristics of Jupiter's moons and to estimate Jupiter's mass using Kepler's Third Law.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Import the Moons class from the jupiter.py module from the moonanalysis  package
from moonanalysis.jupiter import Moons, EnhancedMoons

# Instantiate the class
moons = EnhancedMoons('jupiter.db')
```

## Data Loading and Initial Exploration

First, let's load the data using the `Moons` class and take a preliminary look at the dataset.


```python
# Loading the data
moons_data = moons.data

# Displaying the first few rows of the dataset
moons_data.head()
```

## Kepler's Third Law Analysis for Estimating Jupiter's Mass

Using Kepler's Third Law, we can estimate the mass of Jupiter. This involves preparing the data, training a linear regression model, and then using the model to estimate Jupiter's mass.


```python
# Preparing data for Kepler's Third Law analysis
moons.prepare_kepler_data()

# Setting up and training the linear regression model
moons.setup_linear_model()
training_score = moons.train_model()
print(f"Model Training Score: {training_score}")

# Estimating Jupiter's mass
estimated_mass = moons.predict_mass_of_jupiter()
print(f"Estimated Mass of Jupiter: {estimated_mass} kg")
```

## Comparison with Known Values

Let's compare our estimated mass of Jupiter with the known value from literature.


```python
# Known mass of Jupiter from literature (in kg)
known_mass_jupiter = 1.898e27

# Comparison
estimated_mass, percent_difference = moons.compare_with_literature(known_mass_jupiter)
print(f"Estimated Mass: {estimated_mass} kg")
print(f"Difference from Known Value: {percent_difference}%")
```
